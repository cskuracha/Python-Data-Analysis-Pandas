```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Data/world_population_data.csv")
data
```

```python
country = data['country']
population_2023 = data['2023 population']
```

```python
type(country)
```

```python
population_2023
```

```python
plt.bar(country[:5], population_2023[:5], color = ['red','orange','green','blue','yellow'])
plt.xlabel("Country")
plt.ylabel("2023 Population")
plt.title("Top 5 Populated Countries")
plt.show()
```

```python
plt.bar(country[:5], population_2023[:5], color = ['red','orange','green','blue','yellow'])
plt.grid(axis = 'y',linestyle = "--", alpha = 0.7)
plt.xlabel("Country")
plt.ylabel("2023 Population")
plt.title("Top 5 Populated Countries")
plt.show()
```

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.barplot(data, x = country[:5], y = population_2023[:5], hue = population_2023[:5], palette = sns.color_palette())
plt.xlabel("Country")
plt.ylabel("2023 Population")
plt.title("Top 5 Populated Countries")
plt.show()
```

```python
sns.barplot(x = country[:5], y = population_2023[:5], palette = "viridis")
plt.xlabel("Country")
plt.ylabel("2023 Population")
plt.title("Top 5 Populated Countries")
plt.show()
```

```python
sns.barplot(x = country[:5], y = population_2023[:5], errorbar = "sd", capsize = 0.2, palette="muted", hue = country[:5])
plt.xlabel("Country")
plt.ylabel("2023 Population")
plt.title("Top 5 Populated Countries")
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 30]
# Advanced features with Seaborn
sns.barplot(x=categories, y=values, errorbar="sd", capsize=0.2, palette="muted", hue = categories)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Advanced Bar Chart with Seaborn')

# Show the chart
plt.show()
```
