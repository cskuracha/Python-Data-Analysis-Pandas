## **Creating Plots with Pandas**


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("Data/air_quality_no2.csv", index_col = 0, parse_dates = True)
air_quality.head()
```

```python
air_quality.plot()
plt.show()
```

```python
# Seperate Plot for Paris
air_quality['station_paris'].plot()
plt.show()
```

```python
# Seperate Plot for London
air_quality['station_london'].plot()
plt.show()
```

## **Pandas Plots**


### To Compare NO2 across two stations


```python
air_quality.plot.scatter(x="station_paris", y = "station_london", alpha = 0.8)
plt.show()
```

### **Plot Functions**

```python
[function_name for function_name in dir(air_quality.plot) if not function_name.startswith("_")]
```

### Box Plot

```python
air_quality.plot.box()
plt.show()
```

### Area Plots

```python
axes = air_quality.plot.area(figsize= (12, 4), subplots = True)
plt.show()
```

```python
fig, axs = plt.subplots(figsize = (12, 4))
air_quality.plot.area(ax = axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("no2_concentration.png")
plt.show()
```
