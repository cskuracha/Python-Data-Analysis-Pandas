## **Creating New Columns using existing columns**



```python
import pandas as pd
```

```python
air_quality = pd.read_csv("Data/air_quality_no2.csv", index_col = 0, parse_dates = True)
air_quality
```

```python
air_quality['london_mg_per_cubic'] = air_quality['station_london'] * 1.882
air_quality.head()
```

```python
air_quality['ratio_paris_antwerp'] = air_quality['station_paris'] / air_quality['station_antwerp']
air_quality.head()
```

```python
air_quality = air_quality.rename(columns = {
    "station_antwerp": "BETR801",
    "station_paris": "FR04014",
    "station_london": "London Westminster"
})
```

```python
air_quality.head()
```

```python
air_quality = air_quality.rename(columns = str.lower)
air_quality.head()
```

### **Long to Wide table format**

```python
no2 = air_quality[air_quality['parameter'] == "no2"]
no2
```

```python
# Getting the first two values from no2
no2_subset = no2.sort_index().groupby(['location']).head(2)
no2_subset
```

```python
# Getting location values in seperate columns
no2_subset.pivot(columns = "location", values = "value")
```

#### Plotting

```python
no2.head()
```

```python
no2.pivot(columns = "location", values = "value").plot()
```

### **pivot table**

```python
air_quality.pivot_table(
    values = "value",
    columns = "parameter",
    index = "location",
    aggfunc = "mean"
)
```

```python
air_quality.pivot_table(
    values = "value",
    columns = "parameter",
    index = "location",
    aggfunc = "median"
)
```

```python
air_quality.pivot_table(
    values = "value",
    columns = "parameter",
    index = "location",
    aggfunc = "mean",
    margins = True
)
```

### **Wide to Long format**

```python
no2.head()
```

```python
no2_pivoted = no2.pivot(
    columns = "location",
    values = "value"
).reset_index()
no2_pivoted.head()
```

```python
no2_long = no2_pivoted.melt(
    id_vars = "date.utc"
)
no2_long.head()
```

```python
no2_long = no2_pivoted.melt(
    id_vars = "date.utc",
    value_name = "no2",
    value_vars = ['BETR801','FR04014','London Westminster'],
    var_name = "id_location"
)
no2_long.head()
```
